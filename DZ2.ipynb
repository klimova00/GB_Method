```python
from pprint import pprint
from lxml import html
import requests
```

```python
url = 'https://lenta.ru/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/40.0.2214.115 Safari/537.36'
}

response = requests.get(url, headers=headers)
```

```python
response.text
```

```python
dom = html.fromstring(response.text)
titles = dom.xpath("//time/../../span/text()")
time = dom.xpath("//span/../div/time/text()")
link = dom.xpath("//time/../../span/../../@href")
```

```python
my_dict = {}
i=0
for title in titles:
  title_list = dom.xpath("//time/../../span/text()")[i]
  time_list = dom.xpath("//span/../div/time/text()")[i]
  link_list = url + dom.xpath("//time/../../span/../../@href")[i]
  my_dict[title] ={
      'time': time_list, 'link': link_list}
  i+=1
print(my_dict)
```
